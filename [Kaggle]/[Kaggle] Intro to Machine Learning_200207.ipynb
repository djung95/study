```python
import pandas as pd
melb_data = pd.read_csv("../input/melbourne-housing-snapshot/melb_data.csv")
```

```python
melb_data
```

```python
#결측값이 들어있는 행 전체 삭제하기
melb_data = melb_data.dropna(axis=0)

#종속변수, 독립변수 설정
y = melb_data.Price 
melb_feature = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', 'Longtitude'] 
X = melb_data[melb_feature]
```

```python
#Train / Validation 나누기
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0) 

#Decision Tree Model
from sklearn.tree import DecisionTreeRegressor 
melb_model = DecisionTreeRegressor(random_state = 0) #불러오기
melb_model.fit(train_X,train_y) #학습
melb_predictions = melb_model.predict(val_X) #예측

#Mean Absolute Error = Actual - Predicted
from sklearn.metrics import mean_absolute_error
melb_mae = mean_absolute_error(melb_predictions,val_y) #Validation Set과 검증
print("Validation MAE when not specifying max_leaf_nodes: {:,.0f}".format(melb_mae))
```

```python
#Optimization / Hyperparameter 설정
melb_model_max100 = DecisionTreeRegressor(max_leaf_nodes = 100, random_state = 0)
melb_model_max100.fit(train_X,train_y)
melb_model_max100_predictions = melb_model_max100.predict(val_X)
melb_model_max100_mae = mean_absolute_error(melb_model_max100_predictions,val_y)
print("Validation MAE for best value of max_leaf_nodes: {:,.0f}".format(melb_model_max100_mae))
```

```python
#Random Forest Model
from sklearn.ensemble import RandomForestRegressor
melb_model_rf = RandomForestRegressor(random_state = 0) #불러오기
melb_model_rf.fit(train_X,train_y)
melb_model_rf_predictions = melb_model_rf.predict(val_X)
melb_model_rf_mae = mean_absolute_error(melb_model_rf_predictions,val_y)
print("Validation MAE for Random Forest Model: {:,.0f}".format(melb_model_rf_mae))
```
