```python
import pandas as pd
melb_data = pd.read_csv("../input/melbourne-housing-snapshot/melb_data.csv")
```

```python
from sklearn.model_selection import train_test_split

# Select subset of predictors
cols_to_use = ['Rooms', 'Distance', 'Landsize', 'BuildingArea', 'YearBuilt']
X = melb_data[cols_to_use]

# Select target
y = melb_data.Price

# Separate data into training and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X, y)
```

```python
from xgboost import XGBRegressor

my_model = XGBRegressor()
my_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error

predictions = my_model.predict(X_valid)
print("Mean Absolute Error: "+ str(mean_absolute_error(predictions, y_valid)))
```

```python
# n_estimators
my_model = XGBRegressor(n_estimators=500)
my_model.fit(X_train, y_train)
```

```python
predictions = my_model.predict(X_valid)
print("Mean Absolute Error: "+ str(mean_absolute_error(predictions, y_valid)))
```

```python
# n_estimators + early_stopping_rounds, eval_set, verbose
my_model = XGBRegressor(n_estimators=500)
my_model.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_valid, y_valid)],
             verbose=False)
```

```python
predictions = my_model.predict(X_valid)
print("Mean Absolute Error: "+ str(mean_absolute_error(predictions, y_valid)))
```

```python
# n_estimators + learning_rate + early_stopping_rounds, eval_set, verbose
my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
my_model.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_valid, y_valid)], 
             verbose=False)
```

```python
predictions = my_model.predict(X_valid)
print("Mean Absolute Error: "+ str(mean_absolute_error(predictions, y_valid)))
```

```python
# n_estimators + learning_rate + n_jobs + early_stopping_rounds + eval_set + verbose
my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05, n_jobs=4)
my_model.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_valid, y_valid)], 
             verbose=False)
```

```python
predictions = my_model.predict(X_valid)
print("Mean Absolute Error: "+ str(mean_absolute_error(predictions, y_valid)))
```
