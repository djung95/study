#### 데이터프레임 인덱스 조작

데이터프레임 인덱스 설정 및 제거
set_index: 기존의 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 설정
reset_index: 기존의 행 인덱스를 제거하고 인덱스를 데이터 열로 추가

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                            np.round(np.random.rand(3, 5),2)]).T,
                            columns=["C1","C2","C3","C4"])
df1
```

```python
df2 = df1.set_index("C1") #기존의 인덱스는 사라짐
df2
```

```python
df2.set_index("C2")
```

```python
df2.reset_index()
```

```python
df2.reset_index(drop=True)
```

#### 다중 인덱스

```python
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.rand(5,4),2),
                  columns=[["A","A","B","B"],
                      ["C1","C2","C1","C2"]])
```

```python
df3
```

```python
df3.columns.names = ["Cidx1","Cidx2"]
df3
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6,4),2),
                  columns=[["A","A","B","B"],
                          ["C","D","C","D"]],
                  index=[["M","M","M","F","F","F"],
                         ["id_" + str(i+1) for i in range(3)]*2])
df4
```

```python
df4.columns.names = ["Cidx1","Cidx2"]
df4.index.names = ["Ridx1","Ridx2"]
df4
```

#### 행 인덱스와 열 인데스 교환

```python
stack() 열 인덱스 -> 행 인덱스로 변환
unstack() 행 인덱스 -> 열 인덱스로 변환
```

```python
df4.stack("Cidx1")
```

```python
df4.stack(1)
```

```python
df4.unstack()
```

```python
df4.unstack("Ridx2")
```

```python
df4.unstack(0)
```

#### 연습 문제 2

```python
df_score3=pd.DataFrame({"반":["A"]*5 +["B"]*5,
                        "번호":["ID0"+ str(i) for i in range(5)]*2,
                        "국어":np.random.randint(70,100,10),
                        "영어":np.random.randint(70,100,10),
                        "수학":np.random.randint(70,100,10)},
                      columns=['반', '번호', '국어', '영어', '수학'])
```

```python
df_score3
```

```python
df_score4=df_score3.set_index(["반","번호"])
```

```python
df_score4
```

```python
df_score4["평균"] = df_score4.mean(axis=1).round(2)
```

```python
df_score4
```

```python
df_score5=df_score3.set_index(["번호","반"]).unstack()
```

```python
df_score5
```

```python
df_score5.loc["평균",:] = df_score5.mean()
```

```python
df_score5
```
