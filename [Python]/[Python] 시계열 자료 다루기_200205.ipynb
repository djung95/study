#### shift 연산

```python
np.random.seed(0)
ts = pd.Series(np.random.randn(4), index=pd.date_range("2018-1-1", periods=4,freq="M"))
ts
```

```python
ts.shift(1)
```

```python
ts.shift(-1)
```

```python
ts.shift(1, freq="M")
```

```python
ts.shift(1, freq="W")
```

#### resample 연산

```python
ts = pd.Series(np.random.randn(100), index=pd.date_range("2018-03-22",periods=100,freq="D"))
ts.tail(20)
```

#### 시간 구간이 작아짐 -->  데이터양이 증가함  --> 업 샘플링
#### 시간 구간이 커짐 -->  데이터양이 감소함  --> 다운 샘플링

```python
ts.resample("W").last()
```

```python
ts.resample('M').first()
```

```python
ts = pd.Series(np.random.randn(60),pd.date_range("2018-01-01",periods=60,freq="T"))
ts.head(60)
```

```python
ts.resample('10T').last()
```

```python
ts.resample('10T',closed="right").last()
```

```python
ts.resample("5T").ohlc()
```

```python
ts.resample('30s').ffill().head(20)
```

```python
ts.resample('30s').bfill().head(20)
```
