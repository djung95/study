#### 연습 문제 2

```python
#1
df_score3 = pd.DataFrame({"반":["A"]*5 + ["B"]*5,
              "번호":["IDO"+str(i) for i in range(1,6)]*2,
              "국어":np.random.randint(50,100,10),
              "영어":np.random.randint(50,100,10),
              "수학":np.random.randint(50,100,10)})
df_score3
```

```python
#2
df_score4 = df_score3.set_index(["반","번호"])
df_score4
```

```python
#3
df_score4["평균"] = round(df_score4.mean(axis=1),2)
df_score4
```

```python
df_score3
```

```python
df_score5=df_score3.set_index(["번호","반"]).unstack()
df_score5
```

```python
df_score5.loc["반 평균",:] = df_score5.mean() #행 추가할 때는 df.loc[column,:]
df_score5
```

#### 연습 문제 4.2.11

```python
first = pd.DataFrame({"매출":np.random.randint(5000,10000,6),
                      "비용":np.random.randint(2500,5000,6)},
                     index=["1월","2월","3월","4월","5월","6월"])
first
```

```python
second = pd.DataFrame({"매출":np.random.randint(5000,10000,6),
                      "비용":np.random.randint(2500,5000,6)},
                     index=["7월","8월","9월","10월","11월","12월"])
second
```

```python
full_year = pd.concat([first,second])
full_year
```

```python
full_year=pd.concat([full_year,pd.DataFrame({"이익":full_year["매출"]-full_year["비용"]})],axis=1)
full_year
```

```python
full_year.loc["총 실적",:]=full_year.sum()
```
